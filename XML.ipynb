```python
#Find score
import xml.etree.ElementTree as etree

def get_attr_number(node):
    count = len(node.attrib)
    for child in node:
        count += get_attr_number(child)
    return count

xml = ""
for _ in range(int(input())):
    xml += input()

tree = etree.ElementTree(etree.fromstring(xml))
root = tree.getroot()
print(get_attr_number(root))
```

```python
#Find maximum depth
#
import xml.etree.ElementTree as etree

maxdepth = 0

def depth(elem, level):
    global maxdepth
    if level == maxdepth:
        maxdepth += 1
    for child in elem:
        depth(child, level + 1)

xml = ""
for _ in range(int(input())):
    xml += input()

tree = etree.ElementTree(etree.fromstring(xml))
depth(tree.getroot(), 0)
print(maxdepth)
```
