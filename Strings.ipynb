```python
#STRING SLIPT AND JOIN
def split_and_join(line):
    # step 1: split on spaces -> ["this", "is", "a", "string"]
    words = line.split()

    # step 2: join with "-" -> "this-is-a-string"
    return "-".join(words)

if __name__ == '__main__':
    line = input()
    result = split_and_join(line)
    print(result)
```

```python
#MUTATIONS
def mutate_string(string, position, character):
    new_string = list(string)
    new_string[position] = character
    return "".join(new_string)

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
#FIND A STRING
def count_substring(string, sub_string):
    count = 0
    for i in range(len(string) - len(sub_string) + 1):
        if string[i:i+len(sub_string)] == sub_string:
            count += 1
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    count = count_substring(string, sub_string)
    print(count)
```

```python
#STRING FORMATTING
def print_formatted(n):
    width = len(bin(n)) - 2  # largeur du binaire le plus long
    for i in range(1, n+1):
        deci = str(i).rjust(width)
        octa = oct(i)[2:].rjust(width)
        hexa = hex(i)[2:].upper().rjust(width)
        bina = bin(i)[2:].rjust(width)
        print(f"{deci} {octa} {hexa} {bina}")
```

```python
#ALPHABET RANGOLI
def print_rangoli(size):
    import string
    alphabet = string.ascii_lowercase

    lines = []
    for i in range(size):
        left = "-".join(alphabet[size-1:i:-1])  # lettres décroissantes
        right = "-".join(alphabet[i:size])      # lettres croissantes
        line = left + (("-" if left else "") + right if left else right)
        lines.append(line.center(4*size-3, "-"))

    # afficher le rangoli
    print("\n".join(lines[::-1] + lines[1:]))
if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
#THE MINION GAME
def minion_game(s):
    vowels = "AEIOU"
    kevin_score = 0
    stuart_score = 0
    n = len(s)

    for i in range(n):
        if s[i] in vowels:
            kevin_score += n - i
        else:
            stuart_score += n - i

    if kevin_score > stuart_score:
        print("Kevin", kevin_score)
    elif stuart_score > kevin_score:
        print("Stuart", stuart_score)
    else:
        print("Draw")
```

```python
#MERGE THE TOOLS

def merge_the_tools(s, k):
    n = len(s)
    for i in range(0, n, k):
        substring = s[i:i+k]
        result = ""
        for c in substring:
            if c not in result:
                result += c
        print(result)
```
