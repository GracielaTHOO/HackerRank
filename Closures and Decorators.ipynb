```python
#Standardize Mobile numbers

def wrapper(f):
    def fun(l):
        # nettoyer : garder seulement les 10 derniers chiffres
        cleaned = ["+91 " + n[-10:-5] + " " + n[-5:] for n in l]
        return f(cleaned)
    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

```

```python
#Decorators 2

def person_lister(f):
    def inner(people):
        # trier par Ã¢ge
        people.sort(key=lambda x: int(x[2]))
        return [f(person) for person in people]
    return inner

@person_lister
def name_format(person):
    return "{} {}".format(person[0], person[1])
```
