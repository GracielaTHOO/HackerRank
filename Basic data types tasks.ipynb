```python
#List comprehensions

if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())

    #list comprehension permet de faire des calculs de listes comme les permutations. On commence par definir la liste dans laquelle on renseigne la structure de la liste puis les variables et enfin les conditions sur celles ci
    result = [
        [i,j,k]
        for i in range(0,x+1)
        for j in range(0,y+1)
        for k in range(0,z+1)
        if i + j + k != n
    ]

    print(result)
```

```python
#Find the runner-Up score!

if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())

    list_ordonnee = sorted(set(arr)) #on enleve les doublons et arrange les nombres par ordre croissant 
    runner_up = list_ordonnee [-2]
    print( runner_up)
    
```

```python
#NESTED LISTS 

if __name__ == '__main__':
    records = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name, score])

    # Recuperer toutes les notes
    scores = sorted({score for _, score in records})

    # Deuxieme plus petite note
    second = scores[1]

    # Etudiants ayant cette note
    result = sorted([name for name, score in records if score == second])

    for name in result:
        print(name)
```

```python
# FIND THE PERCENTAGE

if __name__ == '__main__':
    n = int(input())
    student_marks = {}

    for _ in range(n):
        data = input().split()
        name = data[0]
        scores = list(map(float, data[1:]))
        student_marks[name] = scores

    query_name = input()

    # Calcul de la moyenne
    avg = sum(student_marks[query_name]) / len(student_marks[query_name])

    # Affichage avec 2 decimales
    print(f"{avg:.2f}")
```
