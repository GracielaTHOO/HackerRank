```python
#Write a function

#year % 4 == 0 → divisible par 4
#year % 100 != 0 → sauf les années multiples de 100
#year % 400 == 0 → sauf si divisible par 400, alors c’est bissextile quand même

def is_leap(year):
    leap = False
    if (year % 4 == 0 and year %100 != 0) or  (year %400 == 0 ):
        leap = True
    else :
        return False
    return leap

year = int(input())
print(is_leap(year))
```
