```python
#Introduction to sets

# créer un set
s = set([1, 2, 3, 4, 5])
print(s)  # {1, 2, 3, 4, 5}

# ajouter un élément
s.add(6)

# supprimer un élément (erreur si n'existe pas)
s.remove(3)

# supprimer un élément (sans erreur si n'existe pas)
s.discard(10)

# union et intersection
a = {1, 2, 3}
b = {2, 3, 4}
print(a.union(b))         # {1, 2, 3, 4}
print(a.intersection(b))  # {2, 3}


def average(array):
    s = set(array)  # convertir en set pour supprimer les doublons
    return sum(s)/len(s)

# Exemple
n = int(input())
arr = list(map(int, input().split()))
print(average(arr))
```

```python
# Set.add()

s = {1, 2, 3}
s.add(4)
print(s)  # {1, 2, 3, 4}

s.add(2)
print(s)  # {1, 2, 3, 4} → 2 déjà présent, donc pas ajouté


def set_add():
    n = int(input())
    s = set()

    for _ in range(n):
        cmd, *args = input().split()
        if cmd == "add":
            s.add(int(args[0]))

    print(sum(s))

if __name__ == "__main__":
    set_add()

```

```python
#Set .discard(), .remove() & .pop()

s = {1, 2, 3}
s.discard(2)
print(s)  # {1, 3}

s.discard(5)  # 5 n'existe pas, pas d'erreur
print(s)  # {1, 3}


s = {1, 2, 3}
s.remove(2)
print(s)  # {1, 3}

s.remove(5)  # KeyError ! → 5 n'existe pas


s = {1, 2, 3}
x = s.pop()
print(x)  # 1 (par exemple)
print(s)  # {2, 3} → l'élément retiré n’est plus là

n = int(input())
s = set()

for _ in range(n):
    cmd = input().split()
    if cmd[0] == "add":
        s.add(int(cmd[1]))
    elif cmd[0] == "discard":
        s.discard(int(cmd[1]))
    elif cmd[0] == "remove":
        s.remove(int(cmd[1]))
    elif cmd[0] == "pop":
        s.pop()

print(sum(s))
```

```python
#Set .union() Operation
A = {1, 2, 3}
B = {3, 4, 5}

# Méthode union
C = A.union(B)
print(C)  # {1, 2, 3, 4, 5}

# Avec opérateur |
C = A | B
print(C)  # {1, 2, 3, 4, 5}

# Lire le premier set
n = int(input())
A = set(map(int, input().split()))

# Lire le deuxième set
m = int(input())
B = set(map(int, input().split()))

# Union des deux sets
union_set = A.union(B)

# Afficher le nombre d'éléments uniques
print(len(union_set))

```

```python
#Set .difference() Operation
A = {1, 2, 3, 4}
B = {3, 4, 5}

C = A.difference(B)
print(C)  # {1, 2} → éléments de A qui ne sont pas dans B

# Avec opérateur -
C = A - B
print(C)  # {1, 2}


# Lire le premier set
n = int(input())
A = set(map(int, input().split()))

# Lire le deuxième set
m = int(input())
B = set(map(int, input().split()))

# Différence : éléments dans A mais pas dans B
diff_set = A.difference(B)

# Afficher le nombre d'éléments restants
print(len(diff_set))
```

```python
#Set .symmetric_difference() Operation

A = {1, 2, 3}
B = {2, 3, 4}

C = A.symmetric_difference(B)
print(C)  # {1, 4} → 1 est dans A seulement, 4 dans B seulement

# Avec opérateur ^
C = A ^ B
print(C)  # {1, 4}

# Lire le premier set
m = int(input())
M = set(map(int, input().split()))

# Lire le deuxième set
n = int(input())
N = set(map(int, input().split()))

# Symmetric difference
diff = M.symmetric_difference(N)

# Afficher les éléments triés
for x in sorted(diff):
    print(x)
```

```python
#Set Mutations

A = {1, 2, 3, 4, 5}
B = {2, 3}

# A doit garder uniquement les éléments présents dans A et B
A.intersection_update(B)
print(A)  # {2, 3}


n = int(input())
A = set(map(int, input().split()))
m = int(input())

for _ in range(m):
    op, *nums = input().split()
    B = set(map(int, nums[1:]))
    getattr(A, op)(B)  # appel dynamique de la méthode

print(sum(A))

# Lire le set A
n = int(input())
A = set(map(int, input().split()))

# Nombre d'opérations
m = int(input())

# Appliquer chaque opération
for _ in range(m):
    parts = input().split()
    op = parts[0]
    B = set(map(int, parts[2:]))  # les éléments commencent à l'indice 2

    if op == "update":
        A.update(B)
    elif op == "intersection_update":
        A.intersection_update(B)
    elif op == "difference_update":
        A.difference_update(B)
    elif op == "symmetric_difference_update":
        A.symmetric_difference_update(B)

# Afficher la somme finale
print(sum(A))

```

```python
# Check Subset
A.issubset(B)      # True si A ⊆ B
A <= B              # équivalent

for _ in range(int(input())):
    A = set(map(int, input().split()))
    B = set(map(int, input().split()))
    print(A.issubset(B))

```

```python
#Check Strict Superset

A.issuperset(B)      # True si A ⊇ B
A > B                # strict superset


A = set(map(int, input().split()))
n = int(input())
result = all(A > set(map(int, input().split())) for _ in range(n))
print(result)
```
